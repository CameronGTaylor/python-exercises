```python
from math import tan, sqrt, pi
from numpy import arange

plt.figure(figsize=(12,6))
x1 = range(36)
y1 = [sqrt(x) for x in x1]
plt.subplot(2, 2, 1)
plt.plot(x1, y1)

x2 = range(-36,36)
y2 = [x ** 3 for x in x2]
plt.subplot(2, 2, 2)
plt.plot(x2, y2)

x3 = [i * pi / 32 for i in range(-100,100)]
y3 = [tan(x) for x in x3]
plt.subplot(2, 2, 3)
plt.plot(x3, y3)
plt.xticks(arange(-8*pi, 8*pi, pi))
plt.ylim(-5,5)
plt.xlim(-10,10)

x4 = range(-36,36)
y4 = [2 ** x for x in x4]
plt.subplot(2, 2, 4)
plt.plot(x4, y4)
plt.ylim(0,1200)
plt.xlim(0,15)
```

```python
from math import tan, sqrt, pi
from numpy import arange

plt.figure(figsize=(12,6))
x1 = range(36)
y1 = [sqrt(x) for x in x1]
plot1, = plt.plot(x1, y1)

x2 = range(-36,36)
y2 = [x ** 3 for x in x2]
plot2, = plt.plot(x2, y2)

x3 = [i * pi / 32 for i in range(-100,100)]
y3 = [tan(x) for x in x3]
plot3, = plt.plot(x3, y3)

x4 = range(-36,36)
y4 = [2 ** x for x in x4]
plot4, = plt.plot(x4, y4)

plt.xlim(-5,10)
plt.ylim(-10,15)
plt.legend([plot1,plot2,plot3,plot4],
           ['squareroot x','x cubed','tan(x)','2 to x power'])
```
