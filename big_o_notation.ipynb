## Curves to graph
$y = 0x + 1$ and label the curve "O(1)"
$y = log(x)$ and label the curve "O(log n)"
$y = x$ and label the curve "O(n)"
$y = x * log(x)$ and label it "O(n log n)"

```python
import matplotlib.pyplot as plt
from math import factorial, log
```

```python
x1 = range(0,30)
y1 = [0*x + 1 for x in x1]
plt.plot(x1,y1,label='O(1)')

x2 = [x/10 for x in range(1,300)]
y2 = [log(x) for x in x2]
plt.plot(x2,y2,label='O(log n)')

y3 = [x for x in x1]
plt.plot(x1,y3,label='O(n)')

y4 = [x*log(x) for x in x2]
plt.plot(x2,y4,label='O(n log n)')

plt.suptitle('Big O Notation')
plt.xlabel('Elements')
plt.ylabel('Operations')
plt.legend()
plt.ylim(-5,30)
plt.xlim(-5,30)
```

$y = x^2$ and label it "O(n^2)"
$y = 2^x$ and label it "O(2^n)"
$y = n!$ and label is "O(n!)"
$y = n^n$ and label it "O(n^n)"

```python
x1 = [x/10 for x in range(0,300)]
y1 = [x**2 for x in x1]
plt.plot(x1,y1,label='O(n^2)')

y2 = [2**x for x in x1]
plt.plot(x1,y2,label='O(2^n)')

x2 = range(0,30)
y3 = [factorial(x) for x in x2]
plt.plot(x2,y3,label='O(n!)')

y4 = [x**x for x in x1]
plt.plot(x1,y4,label='O(n^n)')

plt.suptitle('Big O Notation')
plt.xlabel('Elements')
plt.ylabel('Operations')
plt.legend()
plt.ylim(0,300)
plt.xlim(0,15)
```

```python
import numpy as np
plt.figure(figsize=(15,9))
plt.style.use('fivethirtyeight')

x = np.linspace(0,100,1000) #range works
x1 = [x/10 for x in range(1,300)] 
x2 = range(0,100)

operations = [0*x+1, x, x**2, 2**x, x**x,]
labels = ['O(1)', 'O(n)', 'O(n^2)', 'O(2^n)', 'O(n^n)',]

for num in range(len(operations)):
    plt.plot(x, operations[num], label=labels[num])

y2 = [log(x) for x in x1]
plt.plot(x1,y2,label='O(log n)')

y4 = [x*log(x) for x in x1]
plt.plot(x1,y4,label='O(n log n)')

y7 = [factorial(x) for x in x2]
plt.plot(x2,y7,label='O(n!)')

plt.suptitle('Big O Notation')
plt.xlabel('Elements')
plt.ylabel('Operations')
plt.legend(loc='upper right')
plt.ylim(-5,40)
plt.xlim(0,8)
```
